```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
%matplotlib inline
```

**Loading the given dataset.**

`



```python
columns=['Sepal Length','Sepal Width','Petal Length','Petal Width','Class Labels' ]
df=pd.read_csv('iris.data',names=columns)
df.head(50);
```

**Visualization of Dateset**

```python
#description of dataset
df.describe()
```

```python
#blue flower has the highest sepal length and is easily distinguishable
#this is visualisation of whole dataset
sn.pairplot(df,hue='Class Labels')
```

**Separating the given data into testing and training**

```python
#Separate the data using slicing of matrices operation
#seapration the data into input and output columns i.e attributes and class labels
data=df.values
X=data[:,0:4]
Y=data[:,4]
print(Y)
```

```python
#split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
print(X_train)  
```

**PREDICTION MODELS**

Model 1 : SUPPORT VECTOR MACHINE ALGORITHM

```python
# Support vector machine algorithm training code
from sklearn.svm import SVC
model_svc=SVC();
model_svc.fit(X_train,Y_train)
```

```python
#prediction values from svc algortihm
predictionvalues1=model_svc.predict(X_test)
#Calculate the Accuracy
from sklearn.metrics import accuracy_score
print( accuracy_score(Y_test,predictionvalues1) * 100)
#percentage of accuracy
#for 45 test data entries
for i in range(len(predictionvalues1)):
  print(Y_test[i],predictionvalues1[i])
# AROUND 1 ENTRY DID NOT MATCH OUT OF 45 TEST DATA ENTRIES


```

Model 2 : Logistic Regression

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)

```

```python
predictionvalues2=model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print( accuracy_score(Y_test,predictionvalues2) * 100)
for i in range(len(predictionvalues2)):
  print(Y_test[i],predictionvalues2[i])
```

Model 3: Decision Tree Classifier 

```python
# decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model_dtc=DecisionTreeClassifier()
model_dtc.fit(X_train,Y_train)
```

```python
predictionvalues3=model_dtc.predict(X_test)
#calculate the accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictionvalues3) * 100)
```

**A detailed clasification report**


```python
# classicifation report for prediction 2
from sklearn.metrics import classification_report
print(classification_report(Y_test,predictionvalues2))
```
